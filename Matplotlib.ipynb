*fig will generate the canvas on which we prepare the graph
since till is shows the only cancva because we had not print anything yet so to add something we need to add 
axes
and while providing axes we need to pass the 4 value those values are:
1st value=left se kitna move krna hai
2nd =bottom se kitna move krna hai
3rd=height
4th=width

and we need to add the values between 0 and 1
so to pass the value we need to pass the decimal value
*

```python
fi=plt.figure()
axis=fi.add_axes([0.1,0.2,1,1])
x=np.array([2,4,2,1,423,4])
y=np.array([42,42,5,252,23,532])
axis.plot(x,y)
axis2=fi.add_axes([0.9,1,0.2,0.2])
axis2.plot(x,y)
```

*since in this fig axis2 we move the fig 50% from bottom and 30% from left and here it shows the proper image
since if we move it 0.9 from left and 1 from bottom then it will looks like this
hence in fig we have complete access to the plot where as in subplot we have to plot the graph in a particular space 
*


**since in figure object we can manage the pixcels using the figsize in which we need to pass the tuple of 2 values**
*by using the figsize we can manage the size of the canvas we are generating* 
**since to manage the pixcels we use dpi - depth per inches **

```python

di=plt.figure(figsize=(2,2),dpi=100)
z=di.add_axes([0.1,.2,1,1])
z.plot(x,y)
```
