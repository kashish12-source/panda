**jointplot**

```python
 # when we want to compare the total bills and the tip so at that time we use the joint plot
sns.jointplot(x="total_bill",y="tip",data=x,kind="scatter")
```

```python
sns.jointplot(x="total_bill",y="tip",data=x,kind="hex")
```

```python
sns.jointplot(x="total_bill",y="tip",data=x,kind="reg")
```

```python
sns.jointplot(x="total_bill",y="tip",data=x,kind="hist")
# ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid'], but 'heg' was passe
```

```python
sns.jointplot(x="total_bill",y="tip",data=x,kind="kde")
```

```python
sns.jointplot(x="total_bill",y="tip",data=x,kind="resid")
```

**pairplot**

```python
sns.pairplot(x ,hue="sex",palette="flare")
# hue will show the diffrence and palette will decide the the color
# this will show the qualitative diffrence
```

```python
sns.pairplot(x,hue="time",palette="pastel")
```

```python
sns.pairplot(x,hue="size",palette="rainbow")
```
