```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

**for categorial plot we use the seaborn**

boxplot
violinplot
stripplot
swarmplot
barplot
countplot


```python
x=sns.load_dataset("tips")
```

```python
sns.boxplot(x["sex"])
```

```python
sns.countplot(x['smoker'])
# since this is printing on the y axis so if you want to print it in x axis you have to define it
```

```python
sns.countplot(x=x['smoker'])
```

```python
# since if you want to classfiy that how many male and female can smoke so for this we use hue:
sns.countplot(x=x['smoker'],hue=x['sex'])
```

**barplot**

```python
sns.barplot(x=x["sex"],y=x['total_bill'] ,color='brown')
```

```python
# since in this plot the average was taken to represent the data 
# this is done for the comparison
sns.barplot(x=x['sex'],y=x['tip'],estimator=np.median)
```

```python
sns.barplot(x=x['sex'],y=x['tip'],estimator=np.max)
```

```python
sns.barplot(x=x['sex'],y=x['tip'],estimator=np.min)
```

```python
sns.barplot(x=x['sex'],y=x['total_bill'],estimator=np.sum)
```

**boxplot**

```python
sns.boxplot(x='tip',y='day',data=x)
```

```python
sns.boxplot(x='total_bill',y='sex',data=x)
```

**violinplot**

```python
sns.violinplot(x='total_bill',y='day',data=x)
```

```python
sns.violinplot(x='sex',y='tip',data=x)
```

**stripplot and swamplot**

```python
sns.stripplot(x='tip',y='sex',data=x)
```
