**regression map**
- for regression map we use the lmplot()
- the regression map is generally used in machine learning
- in regression plot you just have to find the best fit line through the scatter plot and then you can create the regresssion plot
-*to get to know about the details of the methods we use the short cut i.e shift + tab*

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.lmplot(tips,x="total_bill",y='tip')# generaly y-axis contains the value you want to predict and x=axis contains the vaues you know
```

```python
sns.lmplot(tips,x="total_bill",y='tip',hue='sex')
```

```python
sns.lmplot(tips,x="total_bill",y='tip',hue='sex',palette='coolwarm')
```

```python
sns.lmplot(tips,x="total_bill",y='tip',hue='sex',palette='rainbow',markers=['o','v'])
```
